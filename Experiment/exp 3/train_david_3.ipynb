```python
# =========================
# PATHS
# =========================
base_dir = "final_dataset"  # @param {type:"string"}
metadata_csv = "final_dataset/metadata_clean_v2.csv"  # @param {type:"string"}
output_dir = "outputs"  # @param {type:"string"}
best_model_save_path = "outputs/best_detector_model.pt"  # @param {type:"string"}
clean_csv = "outputs/metadata_clean.csv"  # @param {type:"string"}
vificlip_checkpoint = "k400_clip_complete_finetuned_30_epochs.pth"  # @param {type:"string"}

# =========================
# TRAINING
# =========================
learning_rate = 1e-3  # @param {type:"number"}
batch_size = 128  # @param {type:"integer"}
num_epochs = 5  # @param {type:"integer"}
num_frames = 16  # @param {type:"slider", min:8, max:64, step:8}
num_workers = 4  # @param {type:"integer"}
input_dim = 512  # @param {type:"integer"}
num_classes = 1  # @param {type:"integer"}
use_amp = True  # @param {type:"boolean"}
lr_scheduler = "cosine"  # @param ["cosine", "plateau", "none"]
patience = 3  # @param {type:"integer"}

# =========================
# DATASET
# =========================
val_split = 0.2  # @param {type:"number"}
train_size = "30000"  # @param {type:"string"}
val_size = "2000"  # @param {type:"string"}
balance = True  # @param {type:"boolean"}
min_seg_secs = 5  # @param {type:"integer"}
max_seg_secs = 10  # @param {type:"integer"}

# =========================
# WANDB
# =========================
wandb_enabled = True  # @param {type:"boolean"}
wandb_project = "david-deepfake"  # @param {type:"string"}
wandb_run_name = "null"  # @param {type:"string"}


# =========================
# HELPER (handle null)
# =========================
def to_yaml_val(v):
    if isinstance(v, str) and v.lower() == "null":
        return "null"
    if isinstance(v, bool):
        return str(v).lower()
    return v


# =========================
# GENERATE YAML
# =========================
config = f"""
paths:
  base_dir: {base_dir}
  metadata_csv: {metadata_csv}
  output_dir: {output_dir}
  best_model_save_path: {best_model_save_path}
  clean_csv: {clean_csv}
  vificlip_checkpoint: {vificlip_checkpoint}

training:
  learning_rate: {learning_rate}
  batch_size: {batch_size}
  num_epochs: {num_epochs}
  num_frames: {num_frames}
  num_workers: {num_workers}
  input_dim: {input_dim}
  num_classes: {num_classes}
  use_amp: {to_yaml_val(use_amp)}
  lr_scheduler: {lr_scheduler if lr_scheduler != "none" else "null"}
  patience: {to_yaml_val(patience)}

dataset:
  val_split: {val_split}
  train_size: {to_yaml_val(train_size)}
  val_size: {to_yaml_val(val_size)}
  balance: {to_yaml_val(balance)}
  min_seg_secs: {min_seg_secs}
  max_seg_secs: {max_seg_secs}

wandb:
  enabled: {to_yaml_val(wandb_enabled)}
  project: {wandb_project}
  run_name: {to_yaml_val(wandb_run_name)}
"""

with open("config.yaml", "w") as f:
    f.write(config)

print("✅ config.yaml generated!")
```

```python
from config import CLIP_ARCH, CLASS_NAMES, VIFICLIP_CHECKPOINT
from encoder import load_feature_extractor

feature_extractor = load_feature_extractor(
    arch=CLIP_ARCH,
    class_names=CLASS_NAMES,
    checkpoint_path=VIFICLIP_CHECKPOINT,
).cuda()
```

```python
import os
from config import METADATA_CSV, BASE_DIR, BATCH_SIZE, VAL_SPLIT, NUM_WORKERS
from config import TRAIN_SIZE, VAL_SIZE, BALANCE, NUM_FRAMES, CLEAN_CSV, OUTPUT_DIR
from config import INPUT_DIM, NUM_CLASSES, LEARNING_RATE, NUM_EPOCHS, BEST_MODEL_SAVE_PATH, HEAD_TYPE
from config import MIN_SEG_SECS, MAX_SEG_SECS, SPLIT_CACHE
```

```python
from model import build_transform
from dataset import get_train_val_loaders
from train import set_seed

set_seed(42)

preprocess = build_transform(224)
train_loader, val_loader, val_files = get_train_val_loaders(
    transform=preprocess,
    csv_path=METADATA_CSV,
    base_dir=BASE_DIR,
    val_split=VAL_SPLIT,
    train_size=TRAIN_SIZE,
    val_size=VAL_SIZE,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKERS,
    num_frames=NUM_FRAMES,
    validate=False,
    clean_csv_path=CLEAN_CSV,
    balance=BALANCE,
    min_seg_secs=MIN_SEG_SECS,
    max_seg_secs=MAX_SEG_SECS,
    split_cache_path=SPLIT_CACHE,
)
```

```python
from config import USE_AMP, LR_SCHEDULER, PATIENCE
from config import WANDB_ENABLED, WANDB_PROJECT, WANDB_RUN_NAME
from config import CLIP_ARCH, CLASS_NAMES
from train import set_seed, run_training

set_seed(42)

classifier = run_training(
    feature_extractor=feature_extractor,
    train_loader=train_loader,
    val_loader=val_loader,
    input_dim=INPUT_DIM,
    num_classes=NUM_CLASSES,
    head_type=HEAD_TYPE,
    lr=LEARNING_RATE,
    num_epochs=3,
    save_path=BEST_MODEL_SAVE_PATH,
    use_amp=USE_AMP,
    lr_scheduler=LR_SCHEDULER,
    patience=PATIENCE,
    use_wandb=WANDB_ENABLED,
    wandb_project=WANDB_PROJECT,
    wandb_run_name=WANDB_RUN_NAME,
    wandb_extra_config={
        "clip_arch":   CLIP_ARCH,
        "class_names": list(CLASS_NAMES),
        "num_frames":  NUM_FRAMES,
        "balance":     BALANCE,
    },
)
```
